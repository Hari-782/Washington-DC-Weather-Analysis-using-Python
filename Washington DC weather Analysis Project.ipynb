```python
# Import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Insert and read the csv file
data = pd.read_csv('F:/DATA SCIENCE/Datasets/dc_weather.csv')
data
```

```python
# Using head function to get first few values
data.head()
```

```python
data['uvindex'].value_counts()
```

```python
index = [5, 8, 6, 7, 4, 9, 1, 3, 2, 10, 0]
plt.pie(data['uvindex'].value_counts(), autopct = '%1.1f%%', labels = (index), radius = 1.5)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['uvindex'], bins=len(data['uvindex'].unique()), color='skyblue', edgecolor='black')
plt.xlabel('UV Index')
plt.ylabel('Frequency')
plt.title('Histogram of UV Index')
plt.show()
```

```python
sns.displot(data['feelslike'], kde=True)
plt.show()
```

```python
sns.scatterplot(data=data, x='tempmax', y='tempmin')
plt.title('Scatterplot of Max vs Min Temperature')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=data, x='datetime', y='temp')
plt.title('Boxplot of Temperature over Time')
plt.show()
```

```python
sns.pairplot(data, vars=['temp', 'tempmax', 'tempmin', 'feelslike'], hue='uvindex')
plt.show()
```

```python
x = ['temp', 'tempmax', 'tempmin', 'feelslike', 'uvindex']
correlation = data[x].corr()
correlation
```

```python
plt.figure(figsize = (10, 6))
sns.heatmap(correlation, annot = True, cmap = 'crest', fmt = ".5f", square = True)
plt.title('Correlation of Matrix')
plt.show()
```
